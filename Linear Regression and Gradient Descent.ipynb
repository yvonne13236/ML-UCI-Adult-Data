Cost Function

```python
#cost function
def cost_function(X, Y, B):
    m = len(Y)
    J = np.sum((X.dot(B) - Y) ** 2)/(2 * m)
    return J
```

Gradient Descent Function

```python
#implement GD
def gradient_descent(X,Y,B,alpha,iterations):
    cost_history=[0]*iterations
    m=len(Y)
    
    for iteration in range(iterations):
        #Hypothesis Values
        h=X.dot(B)
        #Difference between hypothesis and the actual Y
        loss = h-Y
        #Gradient Calculation
        gradient =X.T.dot(loss)/m
        #changing values of B using Gradient
        B=B-alpha*gradient
        #New cost value
        cost=cost_function(X,Y,B)
        
        cost_history[iteration] = cost #add a new cost the the cost_history list
    
    return B,cost_history
```

Experiment 2 Train

```python
def converged(X,Y,B,alpha,t, max_iter):
    converged = False
    itert = 0
    m=len(Y) # number of samples
    currentcost=cost_function(X,Y,B)
    y_exp2_itert2=[]
    x_exp2b=[]
    
    
    # Iterate Loop
    while not converged:
        #Hypothesis Values
        h=X.dot(B)
        #Difference between hypothesis and the actual Y
        loss = h-Y
        #Gradient Calculation
        gradient =X.T.dot(loss)/m
        #changing values of B using Gradient
        B=B-alpha*gradient
        #New cost value
        newcost=cost_function(X,Y,B)
        
       # cost_history.append(cost) #add a new cost the the cost_history list

        if abs(newcost-currentcost) <= t:
            print ('Converged, iterations: ', itert)
            converged = True
            break
            
        currentcost=newcost 
        itert += 1  # update iter
        
        y_exp2_itert2.append(itert)
        x_exp2b.append(currentcost)
        
        if itert == max_iter:
            print ('Max iterations exceeded!')
            converged = True
        
    
    #Plot how cost changes with iteraitons 
    plt.title("Iterations vs.  Error ")
    plt.xlabel("Iteraitons")
    plt.ylabel(" Error")
    plt.plot(x_exp2b,y_exp2_itert2)
    plt.show()

    return B,newcost,itert

    
```
